# Represent the data using the parallel coordinates method

This method allows to show a set of points described by $n$ variables. Each variable is represented as a vertical lines, parallel to the other variables.
The $n$ parallel lines representing the variables are equally spaced.
The coordinates of each point along variable $i$ is reported in the corresponding line, and then the different coordinates of the point are connected by lines.

Therefore, each point is represented as a $polyline$, which vertices on the parallel lines corresponding to the variables.

https://pandas.pydata.org/pandas-docs/stable/visualization.html

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(df,'species', alpha=0.2)
plt.show()
```
